```python
import pandas as pd
import numpy as np

# 100 random numbers on [0, 5)
array = np.random.randint(0,5,100)

# reshape to 25 rows with 4 columns
reshaped = array.reshape(25,4)

df = pd.DataFrame(reshaped)
df.columns = ['colname_' + str(x) for x in range(4)]

# the cool stuff
# pass a column to get_dummies and the items in the column will each
# receive a new column in a new dataframe where the column
# contains only 0,1 showing the presence of that item in the given row.
dummy_cols = pd.get_dummies(df['colname_3'], prefix='binary_col3')

# Join them, and see the correspondence of the dummy_cols to the
# source column col_3
new_df = df.join(dummy_cols)
print(new_df)
```
